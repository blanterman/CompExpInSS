# 1.7 Transformations of the Time Index for Continuous-Time Signals (S)

## Heaviside Function

```python
# Heaviside Unit Step function
# Returns a vector the same size as the
# input vector, where each element is 1 if the 
# corresponding element of t is greater than or 
# equal to zero.
def Heaviside(t):
    return [1 if val >= 0 else 0 for val in t]

import numpy as np

t = np.linspace(-1, 1, 11)
print(t)
print(Heaviside(t))
```

## Intermediate Problems

```python
# Sympy is the symbolic package I will use for symbollic math
from sympy import *
```

### (a).

```python
# Define the symbols of f(t)
t = symbols('t')

# Define the symbolic function f(t) using the unit step function above
f = t * (Heaviside(t) - Heaviside(t - 2))

# Define the minimum t and maximum t for the plots
mint = -6
maxt = 6

# Plotting the function f(t) from t in range defined above
pf = plot(f, (t, mint, maxt))
```

### (b).

#### $g_1(t) = f(-t)$ 

Should be $f(t)$ flipped

```python
g1t = f.subs([(t, -t)])
pg1 = plot(g1t, (t, mint, maxt))
```

#### $g_2(t) = f(t+1)$

Should be $f(t)$ advanced by 1

```python
g2t = f.subs([(t, t + 1)])
pg2 = plot(g2t, (t, mint, maxt))
```

#### $g_3(t) = f(t - 3)$

Should be $f(t)$ delayed by 3

```python
g3t = f.subs([(t, t - 3)])
pg3 = plot(g3t, (t, mint, maxt))
```

#### $g_4(t) = f(-t + 1)$

Should be $f(t)$ advanced by 1 then flipped

```python
g4t = f.subs([(t, -t + 1)])
pg4 = plot(g4t, (t, mint, maxt))
```

#### $g_5(t) = f(-2t + 1)$

Should be $f(t)$ advanced by 1 then flipped and  linearly compressed by a factor of 1/2.

```python
g5t = f.subs([(t, -2 * t + 1)])
pg5 = plot(g5t, (t, mint, maxt))
```
