```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
n = np.arange(-3, 4)
x = 2 * n
print(x)
plt.stem(n, x)
plt.show()
```

```python
n = np.arange(-5, 6)
x = np.pad(x, (2, 2))  # Pad an array with 2 zeros on left and 2 on right
print(x)
plt.stem(n, x)
plt.show()
```

```python
n = np.arange(-100, 101)
x1 = np.pad(x, (95, 95))  # Pad an array with 95 zeros on left and 95 on right
plt.figure(figsize=(16,4))
plt.stem(n, x1)
plt.show()
```

```python
x = np.concatenate( (np.zeros(95), x, np.zeros(95)))  # Alternate way to pad with zeros
plt.figure(figsize=(16,4))
plt.stem(n, x)
plt.show()
```

```python
## define delta[n]
nx1 = np.arange(0, 11)
x1 = np.concatenate((np.array([1]), np.zeros(10)))
plt.figure(figsize=(16,4))
plt.stem(nx1, x1)
plt.show()
```

```python
## define delta[n+2]
nx2 = np.arange(-5, 6)
x2 = np.concatenate((np.zeros(3), np.array([1]), np.zeros(7)))
## If you plot just plt.stem(x2) you get delta[n-3], so using nx2 in plot is important
plt.figure(figsize=(16,4))
plt.stem(nx2, x2)
plt.show()
```

```python
## Continuous time signals done with symbolic math library SymPy (later)
## Or really close values of t graphed to look continuous
## Each of np.arange(start, end - 1 step, stepsize) and np.linspace(start, stop, numsteps)
## has its advantages
t = np.arange(-5., 5.1, 0.1)
print(t)
t = np.linspace(-5., 5., 101)
print(t)

## Sinusoidal signals
x = np.sin(np.pi * t / 4.)
plt.figure(figsize=(16,4))
plt.plot(t,x) # Use plt.plot to connect dots
plt.show()
```

```python
## Discrete time sinusoids and complex exponentials
## Representing e^(j*pi*n/8) for 0 <= n <= 32
n = np.arange(0, 33)
x = np.exp(1j * (np.pi / 8) * n)
plt.stem(n, np.real(x))
plt.show()
plt.stem(n, np.imag(x))
plt.show()
plt.stem(n, np.absolute(x))
plt.show()
plt.stem(n, np.angle(x)) # This is in radians
plt.show()
plt.stem(n, np.angle(x, deg=True)) # This is in degrees
plt.show()
plt.stem(n, np.angle(x) * 180. / np.pi) # This is also in degrees
plt.show()
```

```python
## Adding, subtracting, multiplying, dividing, scaling, and exponentiating
x1 = np.sin((np.pi / 4.) * np.arange(0, 16))
x2 = np.cos((np.pi / 7.) * np.arange(0, 16))
plt.figure(figsize=(8,64))
plt.subplot(8,1,1)
plt.stem(x1)
plt.subplot(8,1,2)
plt.stem(x2)
## Term by term operations:
## Addition
y1 = x1 + x2  # could also use np.add(x1, x2)
plt.subplot(8,1,3)
plt.stem(y1)
## Subtraction
y2 = x1 - x2  # could also use np.subtract(x1, x2)
plt.subplot(8,1,4)
plt.stem(y2)
## Multiplication
y3 = x1 * x2  # could also use np.multiply(x1, x2)
plt.subplot(8,1,5)
plt.stem(y3)
## Division
y4 = x1 / x2  # could also use np.divide(x1, x2)
plt.subplot(8,1,6)
plt.stem(y4)
## Scaling
y5 = 2 * x1  # could also use np.multiply(2, x1)
plt.subplot(8,1,7)
plt.stem(y5)
## Exponentiation
y6 = np.power(x1, 3)
plt.subplot(8,1,8)
plt.stem(y6)
plt.show()
```

```python
## Labeling and titling
n = np.arange(0, 33)
x = np.exp(1j * (np.pi / 8) * n)
plt.stem(n,np.angle(x))
plt.title('Phase of exp(j*(pi/8)*n)')
plt.xlabel('n (samples)')
plt.ylabel('Phase of x[n] (radians)')
plt.show()
```

```python
# y,z = foo(x) accepts a numerical argument x and returns two
# arguments y and z, where y is 2*x and z is (5/9)*(x-32)
def foo(x):
    y = 2 * x
    z = (5 / 9) * (x - 32)
    return y, z
```

```python
## calling a function
y, z = foo(-40)
print(y, z)
y, z = foo(212)
print(y, z)
```
