## Intermediate Problems

For a continuous-time signal $x(t)$, the energy over the interval $-a \leq t \leq a$ is often defined as

$E_a = \int_{-a}^a \mathrm{|x(t)|}^2\,\mathrm{d}t$, $a \geq 0$

Where $|x|^2=xx^*$ and $x^*$ is the complex conjugate of $x$. Thus for a periodic signal with fundamental period $T$, $E_{T/2}$ contains the signal energy over one period. The energy in the entire signal is defined as 

$E_{\infty} = \lim\limits_{a \to \infty} E_a$,

if the limit exists. While most signals in practice have finite energy, many of the continuous-time signals used as conceptual tools for signals and systems do not. For example, any periodic signal has infinite energy. For these signals, a more useful measure is average power which is simply energy divided by the length of the time interval. Thus, the time-average power over the interval $-a \leq t \leq a$ is

$P_a = \frac{E_a}{2a}$, $a > 0$

and the time average power of the entire signal is 

$P_{\infty} = \lim\limits_{a \to \infty} P_a$,

if the limit exists. In this problem, you will consider how $P_\infty$ and $E_{T/2}$ are related for periodic signals.

### (c).

```python
# Define the symbolic equation for energy of x1 above 
# Don't need to use the complex conjugate, because the signal is purely real. Just square it.
E1 = integrate(x1**2, (t, -a, a)) # Integrating over t gets rid of t, now the variable (symbol) is a
```

```python
# Define the symbolic equation for energy of x2 above 
# Don't need to use the complex conjugate, because the signal is purely real. Just square it.
E2 = integrate(x2**2, (t, -a, a)) # Integrating over t gets rid of t, now the variable (symbol) is a
```

```python
# Defining the complex conjugate of x3 above to use in energy equation
x3_cc = x3.subs([(I, -I)])
# Defining the symbolic equation for energy of x3 above
E3 = integrate(x3 * x3_cc, (t, -a, a)) # complex number time complex conjugate is purely real
```

```python
# x2 has a period of 10, so T/2 is 5 so a = 5.
print("The energy of x_2(t) over one period is {}".format(E2.subs([(a, 5)])))
# Plotting the energy of x2 above over a from 0 to 30, remember this defines a range
ple2 = plot(E2, (a, 0, 30))
```

```python
# x3 has a period of 6, so T/2 is 3 so a = 3.
print("The energy of x_3(t) over one period is {}".format(E3.subs([(a, 3)])))
# Plotting the energy of x3 above over a from 0 to 30, remember this defines a range
ple3 = plot(E3, (a, 0, 30))
```

### (e).

```python
P1 = simplify(E1 / (2 * a))
print(P1)
plp1 = plot(P1, (a, 0.1, 60))
```

```python
P2 = simplify(E2 / (2 * a))
print(P2)
plp2 = plot(P2, (a, 0.1, 60))
```

As $a$ increases $P_a$ converges on a value.

Estimating $P_\infty$:

P1 and P2 converge on $1/2$.

P3 converges on $2$.
    
Comparing $P_\infty$ to $E_{T/2}/T$:

For E1, E2, and E3 above, $E_{T/2}$ is the energy over one period. $E_{T/2}/T$ is dividing the energy over one period by the length of the period. This is actually equal to the time averaged power $P_a = \frac{E_a}{2a}$ for $a=T/2$, or the time averaged power of the original equations over one period. Because these signals are periodic, the overall average power, $P_\infty$ will converge on the value that is equivalent to the time average power of one period. So $P_\infty = P_{T/2}$. This matches what we see above.

$E_{T/2}/T$ for E1 and E2 are $5/10=1/2$

$E_{T/2}/T$ for E3 is $12/6=2$

$E_{ax1} = \int_{-a}^a \mathrm{|\cos(\pi t / 5)|}^2\,\mathrm{d}t$, $a \geq 0$

$E_{ax1} = \int_{-a}^a \mathrm{\cos^2 (\pi t / 5)}\,\mathrm{d}t$, $a \geq 0$

Energy equation for $x_2(t)$

$E_{ax2} = \int_{-a}^a \mathrm{|\sin(\pi t / 5)|}^2\,\mathrm{d}t$, $a \geq 0$

$E_{ax2} = \int_{-a}^a \mathrm{\sin^2 (\pi t / 5)}\,\mathrm{d}t$, $a \geq 0$

Time average power equation for $x_1(t)$

$P_{ax1} = \frac{E_{ax1}}{2a}$, $a > 0$

$P_{ax1} = \frac{1}{2a}\int_{-a}^a \mathrm{\cos^2 (\pi t / 5)}\,\mathrm{d}t$, $a > 0$

Time average power equation for $x_2(t)$

$P_{ax2} = \frac{E_{ax2}}{2a}$, $a > 0$

$P_{ax2} = \frac{1}{2a}\int_{-a}^a \mathrm{\sin^2 (\pi t / 5)}\,\mathrm{d}t$, $a > 0$

Adding the time average power equations for $x_1(t)$ and $x_2(t)$

$P_{atot} = P_{ax1} + P_{ax2}$

$P_{atot} = \frac{1}{2a}\int_{-a}^a \mathrm{\cos^2 (\pi t / 5)}\,\mathrm{d}t + \frac{1}{2a}\int_{-a}^a \mathrm{\sin^2 (\pi t / 5)}\,\mathrm{d}t$, $a > 0$

$P_{atot} = \frac{1}{2a}\big(\int_{-a}^a \mathrm{\cos^2 (\pi t / 5)}\,\mathrm{d}t + \int_{-a}^a \mathrm{\sin^2 (\pi t / 5)}\,\mathrm{d}t\big)$, $a > 0$

$P_{atot} = \frac{1}{2a}\int_{-a}^a \mathrm{\big(\cos^2 (\pi t / 5) + \sin^2 (\pi t / 5)\big)}\,\mathrm{d}t $, $a > 0$

$P_{atot} = \frac{1}{2a}\int_{-a}^a \mathrm{1}\,\mathrm{d}t $, $a > 0$

$P_{atot} = \frac{1}{2a}\big[t\big]_{-a}^a$, $a > 0$

$P_{atot} = \frac{2a}{2a}$, $a > 0$
