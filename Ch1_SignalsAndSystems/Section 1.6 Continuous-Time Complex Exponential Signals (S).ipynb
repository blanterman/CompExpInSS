```python
# Sympy is the symbolic package I will use for symbollic math
from sympy import *
```

### (a).

```python
# Define the variables in the expression
t, T = symbols('t T')

# Define the expression that contain the symbols
x = sin(2*pi*t/T)

# I wish to plot, but x has two symbols, We will plot for T=5. This creates a new expression x5.
# This substitutes 5 for T in the expression for x
x5 = x.subs([(T,5)])

# Plotting x5 for t from 0 to 10 (two periods)
p1 = plot(x5,(t,0,10),title='$sin(2/5 * \pi /t)$')
```

### (b).

```python
# Defining the symbols for x(t)
t, T = symbols('t T')

# Defining the signal x(t) in terms of T
x = cos(2 * pi * t / T) * sin(2 * pi * t / T)

# Setting x(t) with T=4
x4 = x.subs([(T, 4)])
# Plotting x(t) with T=4 with t from 0 to 32
px4 = plot(x4, (t, 0, 32), title='$cos(2 \pi t / 4)sin(2 \pi t / 4)$')

# Setting x(t) with T=8
x8 = x.subs([(T, 8)])
# Plotting x(t) with T=8 with t from 0 to 32 
px8 = plot(x8, (t, 0, 32), title='$cos(2 \pi t / 8)sin(2 \pi t / 8)$')

# Setting x(t) with T=16
x16 = x.subs([(T, 16)])
# Plotting x(t) with T=16 with t from 0 to 32
px16 = plot(x16, (t, 0, 32), title='$cos(2 \pi t / 16)sin(2 \pi t / 16)$')
```

Based on the graphs above, the fundamental period of $x(t)$ in terms of T is $T/2$.

To see the relationship between Q and the number of periods, we need to calculate Q for all the values of a. Using the equation:

$e^{-at}cos(2 \pi t / T) u(t)$

Q is defined as:

$Q = \frac{2  \pi / T}{2a}$

So for us, $T=0$. We can calculate Q values for all our a values.

```python
# Using numpy to calculate the Q values using a values.
import numpy as np

# a values that were used
a = np.array([0.5, 0.25, 0.125])

# calculaint Q values and printing the results
Q = (2 * np.pi) / (2 * a)
print(a, Q)
```

Here is a table showing the values of a, Q and the number of periods to see if there is a correlation between Q and number of periods.

|a|Q|Num Periods|
|:-:|:-:|:-:|
|0.5|6.28|4|
|0.25|12.57|9|
|0.125|25.13|18|

Based on the Table above. It does indeed appear that the number of periods is proportional to Q.

Why the discontinuities at 8 and 24?

As we see from the magnitude graph, the magnitude of the complex vector goes to zero at 8 and 24 (and presumably every 8+k16 for integer values of k). The interesting thing is when we look at the quadrant that the vector is in before and after the discontinuity. On the real axis x goes from (-) to 0 to (-), whereas on the imaginary axis x goes from (-) to 0 to (+). So before the discontinuity, the vector is in quadrant III, which corresponds to arg values of $\pi$ to $3 \pi / 2$, and after the discontinuity, the vector is in quadrant II, which corresponds to arg values of $\pi / 2$ to $\pi$. This can be seen below in the dynamic graph of the complex vector changing over time. You can see the length go to zero in quadrant III and emerge in quadrant II. This is causing the discontinuity in the angle graph.

```python
# Dynamic plot showing how the vector of the complex equation for x(t) changes over time t from 0 to 32.

# This is the magic command that makes dynamic updating possible in a notebook
%matplotlib notebook

# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# This tells the plot can be dynamically updated
plt.ion()

# Define the x-y coordinates for the complex vector equation x(t) over t from 0 to 32
t = np.linspace(0, 32, 500)
x_coords = np.real(np.exp(1j * 2 * np.pi * t / 16) + np.exp(1j * 2 * np.pi * t / 8))
y_coords = np.imag(np.exp(1j * 2 * np.pi * t / 16) + np.exp(1j * 2 * np.pi * t / 8))

# Creating arrow
x_pos = 0    # setting x value of beginning of arrow
y_pos = 0    # setting y value of beginning of arrow
x_direct = x_coords[0]    # setting x value of end of arrow
y_direct = y_coords[0]    # setting y value of end of arrow
 
# Creating plot
fig, ax = plt.subplots(figsize = (6, 6))

# Setting the x and y limits to show the whole vector
plt.xlim([-2, 2])
plt.ylim([-2, 2])

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Define the vector arrow for the quiver plot and placing on the plot. This will have no scaling
arrow = ax.quiver(x_pos, y_pos, x_direct, y_direct, color='blue',  angles='xy', scale_units='xy', scale=1)
ax.set_title('Quiver plot of x(t)')

# Loop to dynamically update the xy-coordinates of the end of the 
for index in range(len(t)):
    arrow.set_UVC(x_coords[index], y_coords[index])    # update the endpoint
    fig.canvas.draw()    # Redraw the graph
    fig.canvas.flush_events()    # Cleanup
```
