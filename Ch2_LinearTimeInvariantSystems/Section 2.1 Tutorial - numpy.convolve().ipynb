```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Defining the n values for the non-zero values of x[n]
n = np.arange(0,6)

# Defining x[n] non-zero values that correspond to n values above
xn = [1,1,1,1,1,1]

# Will be calculating convolution of x[n] with x[n], the n values for this are calculated using the n values above
# If h[n] wasn't x[n] then we would be using two different n vectors.
ny = np.arange(min(n) + min(n), (min(n) + min(n)) + len(n) + len(n) - 2 + 1)

# Calculate the convolution
y = np.convolve(xn, xn)

# Plot the convolution, marrying the determined n values for the convolution and the convolution itself
plt.stem(ny, y)
plt.title("$y[n] = x[n] * x[n]$")
plt.xlabel('n (samples)')
plt.ylabel("$y[n]$")
plt.show()
```

### (c).
Consider the finite-length signal

Equation (2.6)
<center>
$ 
h[n] =
  \begin{cases}
    n,       & \quad  0 \le n \le 5,\\
    0,  & \quad \text{otherwise}.
  \end{cases}
$
</center>

Analytically determine $y[n] = x[n] * h[n]$. Next, compute y using np.convolve, where your first step should be to define the vector h to contain $h[n]$ on the interval $0 \le n \le 5$. Again construct a vector ny which containg the interval of n for which y contains $y[n]$. Plot your results using plt.stem(ny,y). As a chech, your plot should agree with Figure 2.2 and your analytical derivation.

Analytical derivation done in notebook.

```python
# Defining the n values for x[n] and h[n]
nx = np.arange(0,6)
nh = np.arange(0,6)

# Defining the valuse of x[n] and h[n] corresponding to the values of n defined above
x = np.array([1 for val in nx])
h = np.array([val for val in nh])

# using the values of n for x[n] and h[n] above to get the values of n for the convolution
ny = np.arange(min(nx) + min(nh), min(nx) + min(nh) + len(nx) + len(nh) - 2 + 1)

# Computing the convolution of x[n] and h[n]
y = np.convolve(x, h)

# Plotting x[n], h[n], and y[n]
plt.figure(figsize=(8, 24))
# x[n]
plt.subplot(3,1,1)
plt.stem(nx, x)
plt.xlabel("n")
plt.ylabel("x[n]")
# h[n]
plt.subplot(3,1,2)
plt.stem(nh, h)
plt.xlabel("n")
plt.ylabel("h[n]")
#y[n]
plt.subplot(3,1,3)
plt.stem(ny, y)
plt.xlabel("n")
plt.ylabel("$y[n]=x[n]*h[n]$")

plt.show()
```

### (d). 
How does $y_2[n] = x[n] * h[n+5]$ compare to the signal $y[n] derived in Part (c)?

The values of $y_2[n]$ will be identical to $y[n]$, but the indices applied to those values will be different. The first value will be at $n = -5$ and the last value will be at $n=5$. Advancing $h[n]$ by a factor of 5 advances $y_2[n]$ by a factor of 5.

### (e).
Use np.convolve to compute the nonzero samples of $y_2[n]$, and store these samples in the vector y2. If done correctly, this vector should be identical to the vector y computed in Part (c). The only difference is that the indices of the values sotred in y2 have changed. Determine this set of indices, and store them in the vector ny2 Plot $y_2[n]$ using plt.stem(ny2,y2). Your plot should agree with Figure 2.3.

```python
# Defining the n values for x[n] and h[n] and h[n+5]
nx = np.arange(0,6)
nh = np.arange(0,6)
nhPlus5 = np.arange(-5,1)

# Defining the valuse of x[n] and h[n] corresponding to the values of n defined above
x = np.array([1 for val in nx])
h = np.array([val for val in nh]) # Use nh (original values of nh) to define h[n]. We need an advanced version of this, not a redefinition.

# using the values of n for x[n] and h[n + 5] above to get the values of n for the convolution
ny2 = np.arange(min(nx) + min(nhPlus5), min(nx) + min(nhPlus5) + len(nx) + len(nhPlus5) - 2 + 1)

# Computing the convolution of x[n] and h[n]
y2 = np.convolve(x, h)

# Plotting x[n], h[n+5], and y_2[n]
plt.figure(figsize=(8, 24))
# x[n]
plt.subplot(3,1,1)
plt.stem(nx, x)
plt.xlabel("n")
plt.ylabel("x[n]")
# h[n+5]
plt.subplot(3,1,2)
plt.stem(nhPlus5, h)
plt.xlabel("n")
plt.ylabel("h[n+5]")
#y[n]
plt.subplot(3,1,3)
plt.stem(ny2, y2)
plt.xlabel("n")
plt.ylabel("$y_2[n]=x[n]*h[n+5]$")

plt.show()
```
